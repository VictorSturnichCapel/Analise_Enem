```python
colunasSelecionadas = ['NU_INSCRICAO', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q001', 'Q002']
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=colunasSelecionadas)
```

```python
microdadosEnemSelecionado.head()
```

```python
microdadosEnemSelecionado = microdadosEnemSelecionado.dropna()
```

```python
microdadosEnemSelecionado
```

```python
q001e002Dicionario = {
    'A':'Nunca estudou.',
    'B':'Não completou a 4ª série/5º ano do Ensino Fundamental.',
    'C':'Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental.',
    'D':'Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.',
    'E':'Completou o Ensino Médio, mas não completou a Faculdade.',
    'F':'Completou a Faculdade, mas não completou a Pós-graduação.',
    'G':'Completou a Pós-graduação.',
    'H':'Não sei.',
}
```

```python
microdadosEnemSelecionado.filter(items=['Q001', 'NU_INSCRICAO']).groupby('Q001').count()
```

```python
microdadosEnemSelecionado['NO_Q001'] = [q001e002Dicionario[resp] for resp in  microdadosEnemSelecionado.Q001] 
```

```python
microdadosEnemSelecionado.head()
```

```python
microdadosEnemSelecionado['NO_Q002'] = [q001e002Dicionario[resp] for resp in  microdadosEnemSelecionado.Q002] 
```

```python
microdadosEnemSelecionado.head()
```

```python
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

```python
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q002'])\
                                    .groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

```python
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q001'])\
                        .groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```python
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002')\
                            .mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```python
q002Redacao = microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002')\
                                        .mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```python
q002Redacao.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.suptitle('Nota Redação x Escolaridade X Estado')

microdadosEnemSelecionado.filter(items=['SG_UF_RESIDENCIA', 'NO_Q002', 'NU_NOTA_REDACAO'])\
                        .where(microdadosEnemSelecionado.Q002 !='H')\
                        .groupby(['NO_Q002', 'SG_UF_RESIDENCIA'])\
                        .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)\
                        .unstack().plot(ax=ax, )
```
